```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
print('import successfully')
```

```python
filepath ='Desktop/mi_data.csv'
df = pd.read_csv(filepath,index_col=None)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.head(560)
```

```python
df['Battery Capacity'] = df['battery'].str.extract(r'(\d+)').astype(int)
df.drop('battery', axis=1, inplace=True)
df.head()
```

```python
speed_counts = df['Battery Capacity'].value_counts()
speed_counts = speed_counts.sort_index()
plt.figure(figsize=(10,5))  
plt.plot(speed_counts.index, speed_counts.values, marker='o', color='blue')
plt.xlabel('Battery Capacity')
plt.ylabel('No.of Mobiles')
plt.title('Visualization of Battery Capacity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(True)  
plt.show()
```

```python
df['Network Type'] = df['network'].str.extract(r'Network Type(..)')
df.drop('network', axis=1, inplace=True)
df.head()
```

```python
network_counts = df['Network Type'].value_counts()
plt.figure(figsize=(8,8))  
plt.pie(network_counts, labels=network_counts.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Network Type')
plt.axis('equal')  
plt.show()
```

```python
df['RAM'] = df['storage_ram'].str.extract(r'(\d+)').astype(int)
df.drop('storage_ram', axis=1, inplace=True)
df.head()
```

```python
ram_counts = df['RAM'].value_counts()
plt.figure(figsize=(8, 8))  
plt.pie(ram_counts, labels=ram_counts.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('RAM')
plt.axis('equal') 
plt.show()
```

```python
df['Android Version'] = df['os_processor'].str.extract(r'Android (.*?)Processor')
df['Processor'] = df['os_processor'].str.extract(r'Processor (.*?)Speed')
df['Speed'] = df['os_processor'].str.extract(r'Speed(.*?)GHz')

df.drop('os_processor', axis=1, inplace=True)
```

```python
android_version_counts = df['Android Version'].value_counts()
android_version_counts = android_version_counts.sort_values(ascending=False)
plt.figure(figsize=(10, 6)) 
plt.bar(android_version_counts.index, android_version_counts.values, color='lightblue', edgecolor='black')
plt.xlabel('Android Version')
plt.ylabel('No of Mobiles')
plt.title('Android Versions')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
speed_counts = df['Speed'].value_counts()
speed_counts = speed_counts.sort_index()
plt.figure(figsize=(10, 6))  
plt.plot(speed_counts.index, speed_counts.values, marker='o', color='orange')
plt.xlabel('Processor Speed (GHz)')
plt.ylabel('No.of Mobiles')
plt.title('Processor Speed Graph')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(True)  
plt.show()
```

```python
speed_counts = df['ratings'].value_counts()
speed_counts = speed_counts.sort_index()
plt.figure(figsize=(8, 4))  
plt.plot(speed_counts.index, speed_counts.values, marker='o', color='Blue')
plt.xlabel('Ratings')
plt.ylabel('No.of Mobiles')
plt.title('Ratings Graph')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(True)  
plt.show()
```

```python
df['price'] = df['price'].replace('[^\d.]', '', regex=True)
df['price'] = pd.to_numeric(df['price'])
df['price'] = df['price']
df.head()
```

```python
plt.hist(df['price'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Price')
plt.ylabel('No.of Mobiles')
plt.title('Visualization of Prices')
plt.show()
```
